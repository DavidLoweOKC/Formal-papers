````
def safe_move(src, dst):
    """Move file, creating parent dirs as needed. Handle conflicts."""
    dst = Path(dst)
    dst.parent.mkdir(parents=True, exist_ok=True)
    if dst.exists():
        # Add suffix to avoid overwrite
        stem = dst.stem
        suffix = dst.suffix
        i = 2
        while dst.exists():
            dst = dst.parent / f"{stem}_{i}{suffix}"
            i += 1
    shutil.move(str(src), str(dst))
````

````
def move_tree(src, dst):
    """Move entire directory tree."""
    dst = Path(dst)
    dst.parent.mkdir(parents=True, exist_ok=True)
    if dst.exists():
        # Merge into existing
        for item in Path(src).rglob("*"):
            if item.is_file():
                rel = item.relative_to(src)
                safe_move(item, dst / rel)
        # Remove empty source dirs
        shutil.rmtree(str(src), ignore_errors=True)
    else:
        shutil.move(str(src), str(dst))
````

````
def run_migration():
    moved = 0
    skipped = 0
    errors = []

    # --- Wipe existing partial builds ---
    for d in [BUILD, ARCHIVE]:
        if d.exists():
            shutil.rmtree(str(d))
    BUILD.mkdir()
    ARCHIVE.mkdir()

    print("=" * 60)
    print("PHASE 1: Building _CANONICAL_BUILD")
    print("=" * 60)

    # 01_CORE_FORMALISM
    print("\n>> 01_CORE_FORMALISM")
    for name, src in CORE_FORMALISM.items():
        dst = BUILD / "01_CORE_FORMALISM" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # 02_TEN_LAWS (move entire law folders)
    print("\n>> 02_TEN_LAWS")
    for law_folder in TEN_LAWS_FOLDERS:
        src = ROOT / "01_CANONICAL_EQUATIONS" / law_folder
        dst = BUILD / "02_TEN_LAWS" / law_folder
        if src.exists():
            file_count = sum(1 for _ in src.rglob("*") if _.is_file())
            move_tree(src, dst)
            print(f"   ✓ {law_folder} ({file_count} files)")
            moved += file_count
        else:
            print(f"   ✗ MISSING: {law_folder}")
            skipped += 1

    # 03_FORMAL_MATH
    print("\n>> 03_FORMAL_MATH")
    for name, src in FORMAL_MATH.items():
        dst = BUILD / "03_FORMAL_MATH" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # 04_AXIOM_ENTRIES
    print("\n>> 04_AXIOM_ENTRIES")
    for name, src in AXIOM_ENTRIES.items():
        dst = BUILD / "04_AXIOM_ENTRIES" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # 05_EVIDENCE
    print("\n>> 05_EVIDENCE")
    for name, src in EVIDENCE.items():
        dst = BUILD / "05_EVIDENCE" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # 06_COMPUTATIONAL
    print("\n>> 06_COMPUTATIONAL")
    for name, src in COMPUTATIONAL.items():
        dst = BUILD / "06_COMPUTATIONAL" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # 07_PAPERS
    print("\n>> 07_PAPERS")
    for name, src in PAPERS.items():
        dst = BUILD / "07_PAPERS" / name
        if src.exists():
            safe_move(src, dst)
            print(f"   ✓ {name}")
            moved += 1
        else:
            print(f"   ✗ MISSING: {src}")
            skipped += 1

    # ──────────────────────────────────────────────
    # PHASE 2: Everything remaining → _EXCESS_ARCHIVE
    # ──────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("PHASE 2: Moving remaining files to _EXCESS_ARCHIVE")
    print("=" * 60)

    # Move old numbered folders
    old_folders = [
        "01_CANONICAL_EQUATIONS", "02_FORMAL_MATH", "03_PAPERS_AND_DRAFTS",
        "04_AXIOM_ENTRIES", "05_EVIDENCE_AND_TESTING", "06_DATA_EXPORTS",
        "07_SYSTEM_AND_TOOLS", "08_PROMPTS", "09_DATA_VAULT_ARCHIVE",
        "ZZZ_NUMBERED_DUPLICATES"
    ]
    for folder_name in old_folders:
        src = ROOT / folder_name
        if src.exists():
            # Check if it still has files
            remaining = list(src.rglob("*"))
            remaining_files = [f for f in remaining if f.is_file()]
            if remaining_files:
                dst = ARCHIVE / folder_name
                move_tree(src, dst)
                print(f"   → {folder_name} ({len(remaining_files)} remaining files)")
                moved += len(remaining_files)
            else:
                shutil.rmtree(str(src), ignore_errors=True)
                print(f"   ∅ {folder_name} (empty, removed)")

    # Move loose root files to archive
    print("\n>> Loose root files")
    KEEP_AT_ROOT = {
        "migrate_academic.py",  # this script
    }
    KEEP_DIRS = {"_CANONICAL_BUILD", "_EXCESS_ARCHIVE", "jax_modular", ".git"}

    for item in ROOT.iterdir():
        if item.name in KEEP_AT_ROOT or item.name in KEEP_DIRS:
            continue
        if item.is_file():
            safe_move(item, ARCHIVE / "root_files" / item.name)
            print(f"   → {item.name}")
            moved += 1
        elif item.is_dir() and item.name not in KEEP_DIRS:
            file_count = sum(1 for _ in item.rglob("*") if _.is_file())
            move_tree(item, ARCHIVE / item.name)
            print(f"   → {item.name}/ ({file_count} files)")
            moved += file_count

    print("\n" + "=" * 60)
    print(f"MIGRATION COMPLETE: {moved} files moved, {skipped} skipped")
    print("=" * 60)

    # ──────────────────────────────────────────────
    # PHASE 3: Generate clean INDEX.md
    # ──────────────────────────────────────────────
    generate_index()
    generate_archive_readme()
````
