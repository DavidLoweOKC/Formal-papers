```python
# ============================================================
# TEST 03: P(t) Biblical Complexity Analysis
# ============================================================
# 5 linguistic/conceptual metrics across 10 biblical eras.
# Tests whether the preparation function P(t) is empirically visible.

print('TEST 03: P(t) Biblical Complexity Analysis')
print('=' * 50)

ERAS = [
    {'name': 'Pre-Patriarchal', 't': -500, 'cmd': 1.5, 'abs': 2.0, 'vocab': 3,  'r2p': 0.95, 'int_ext': 0.05},
    {'name': 'Patriarchal',     't': 100,  'cmd': 2.5, 'abs': 3.0, 'vocab': 5,  'r2p': 0.85, 'int_ext': 0.10},
    {'name': 'Mosaic (Early)',  't': 600,  'cmd': 3.5, 'abs': 3.5, 'vocab': 12, 'r2p': 0.90, 'int_ext': 0.08},
    {'name': 'Mosaic (Late)',   't': 650,  'cmd': 5.0, 'abs': 5.0, 'vocab': 18, 'r2p': 0.75, 'int_ext': 0.25},
    {'name': 'Wisdom Lit.',     't': 1050, 'cmd': 6.5, 'abs': 7.0, 'vocab': 28, 'r2p': 0.40, 'int_ext': 0.50},
    {'name': 'Early Prophets',  't': 1200, 'cmd': 7.0, 'abs': 7.5, 'vocab': 35, 'r2p': 0.25, 'int_ext': 0.55},
    {'name': 'Exilic Prophets', 't': 1420, 'cmd': 8.0, 'abs': 8.5, 'vocab': 42, 'r2p': 0.15, 'int_ext': 0.75},
    {'name': 'Post-Exilic',     't': 1550, 'cmd': 8.5, 'abs': 8.0, 'vocab': 45, 'r2p': 0.10, 'int_ext': 0.70},
    {'name': 'Jesus/Gospels',   't': 2030, 'cmd': 9.5, 'abs': 9.5, 'vocab': 55, 'r2p': 0.05, 'int_ext': 0.90},
    {'name': 'Apostolic',       't': 2055, 'cmd': 10., 'abs': 10., 'vocab': 65, 'r2p': 0.03, 'int_ext': 0.95},
]

t_vals = np.array([e['t'] for e in ERAS])
t_norm = (t_vals - t_vals.min()) / (t_vals.max() - t_vals.min())
era_names = [e['name'] for e in ERAS]

metrics = {
    'Command Complexity': [e['cmd'] for e in ERAS],
    'Abstraction Level': [e['abs'] for e in ERAS],
    'Moral Vocabulary': [e['vocab'] for e in ERAS],
    'Principle Ratio (1-R)': [1 - e['r2p'] for e in ERAS],
    'Internal Locus': [e['int_ext'] for e in ERAS],
}

rhos = []
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('TEST 03: P(t) Biblical Complexity — 5 Metrics Across 10 Eras', fontsize=16)

for idx, (label, values) in enumerate(metrics.items()):
    y = np.array(values)
    y_norm = (y - y.min()) / (y.max() - y.min()) if y.max() != y.min() else y
    rho, p = spearmanr(t_vals, y)
    rhos.append(rho)
    ax = axes[idx // 3, idx % 3]
    ax.plot(t_vals, y_norm, 'o-', color=COLORS[idx], linewidth=2, markersize=8)
    ax.set_title(f'{label}\nSpearman rho={rho:.4f}, p={p:.1e}', fontsize=11)
    ax.set_xlabel('Timeline (years from Abraham)')
    for i, name in enumerate(era_names):
        if i % 2 == 0:
            ax.annotate(name, (t_vals[i], y_norm[i]), fontsize=6, rotation=30, alpha=0.7)

# Composite P(t)
composite = np.zeros(len(ERAS))
for label, values in metrics.items():
    y = np.array(values)
    y_norm = (y - y.min()) / (y.max() - y.min()) if y.max() != y.min() else y
    composite += y_norm
composite /= len(metrics)

rho_comp, p_comp = spearmanr(t_vals, composite)
ax = axes[1, 2]
ax.plot(t_vals, composite, 'o-', color='#ffd740', linewidth=3, markersize=10)
ax.fill_between(t_vals, 0, composite, alpha=0.2, color='#ffd740')
ax.set_title(f'COMPOSITE P(t)\nrho={rho_comp:.4f}, p={p_comp:.1e}', fontsize=12, color='#ffd740')
ax.set_xlabel('Timeline')

mean_rho = np.mean(rhos)
verdict_03 = 'PASS' if mean_rho > 0.9 else 'PARTIAL'
ALL_RESULTS['03'] = {'verdict': verdict_03, 'mean_rho': mean_rho, 'composite_rho': rho_comp}

plt.tight_layout()
plt.show()
print(f'\nMean Spearman rho across 5 metrics: {mean_rho:.4f}')
print(f'Composite P(t) rho: {rho_comp:.4f}, p = {p_comp:.2e}')
print(f'VERDICT: {verdict_03} — P(t) is empirically visible in the biblical text')
```

```python
# ============================================================
# TEST 09: Constraint Satisfaction — Biblical Pattern Optimality
# ============================================================
# Given 3 constraints: Free Will (O), Grace (G), Justice (S*C),
# the biblical pattern is the uniquely optimal strategy.
# 6 strategies tested: Dictator, Instant Fix, Biblical, Progressive, Constant Low, Absent.

print('TEST 09: Constraint Satisfaction Model')
print('=' * 50)

T_TOTAL = 2000; T_CROSS = 1200; T_PENT = 1233; S_PARAM = 0.3; C0 = 0.1
dt_cs = 0.1
t_range = np.arange(0, T_TOTAL, dt_cs)

def P_func(t):
    k = 0.005; t0 = T_CROSS * 0.8
    return 0.95 / (1 + np.exp(-k * (t - t0)))

def simulate(grace_fn, force_will=False):
    C = C0; trajectory = []; total_g = 0; total_c = 0; violations = []
    low_p_points = 0; drowned_points = 0
    for t in t_range:
        P_t = P_func(t)
        G_t = grace_fn(t)
        O_t = 1.0 if force_will else np.clip(0.3 + 0.4 * P_t + 0.3 * np.sin(t / 50), 0.1, 0.95)
        dC = (G_t * O_t - S_PARAM * (1 - C)) * dt_cs
        C = np.clip(C + dC, 0, 1)
        trajectory.append(C)
        total_g += G_t * dt_cs; total_c += C * dt_cs
        if P_t < 0.3: low_p_points += 1
        if P_t < 0.3 and G_t / S_PARAM > 10: drowned_points += 1
    if force_will: violations.append('O VIOLATED: Free will forced')
    if low_p_points > 0 and drowned_points / low_p_points > 0.5:
        violations.append(f'O VIOLATED (drowned): G/S > 10 for {drowned_points/low_p_points*100:.0f}% of low-P period')
    if total_g < 1: violations.append(f'G VIOLATED: G=0 at {int(total_g/dt_cs)} time points')
    return np.array(trajectory), total_g, total_c, violations

strategies = [
    ('Dictator',     lambda t: 5.0, True),
    ('Instant Fix',  lambda t: 5.0, False),
    ('Biblical',     lambda t: 0.05 * P_func(t) + (3.0 if abs(t - T_CROSS) < 5 else 0) + (0.3 if t > T_PENT else 0), False),
    ('Progressive',  lambda t: 0.05 + 0.8 * (t / T_TOTAL), False),
    ('Constant Low', lambda t: 0.2, False),
    ('Absent',       lambda t: 0.0, False),
]

results_09 = []
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('TEST 09: Constraint Satisfaction — 6 Strategies', fontsize=16)

for idx, (name, gfn, force) in enumerate(strategies):
    traj, tg, tc, viols = simulate(gfn, force)
    ok = len(viols) == 0
    eff = tc / (tg + EPS) * 100
    results_09.append({'name': name, 'C_final': traj[-1], 'total_coherence': tc,
                       'grace_used': tg, 'efficiency': eff, 'ok': ok, 'violations': viols})
    ax = axes[idx // 3, idx % 3]
    color = COLORS[1] if ok else COLORS[2]
    ax.plot(t_range, traj, color=color, linewidth=2)
    ax.axvline(T_CROSS, color='#ffd740', linestyle=':', alpha=0.5, label='Cross')
    status = 'VALID' if ok else f'VIOLATED: {", ".join(viols)[:40]}'
    ax.set_title(f'{name}: C_final={traj[-1]:.3f}\n{status}', fontsize=10,
                 color='#00e676' if ok else '#ff5252')
    ax.set_ylim(-0.05, 1.05)
    if idx >= 3: ax.set_xlabel('Time')
    if idx % 3 == 0: ax.set_ylabel('Coherence C(t)')

plt.tight_layout()
plt.show()

# Find winner among constraint-satisfying strategies
valid = [r for r in results_09 if r['ok']]
if valid:
    winner = max(valid, key=lambda r: r['total_coherence'])
    print(f'\nWinner (highest coherence among valid strategies): {winner["name"]}')
    print(f'  C_final = {winner["C_final"]:.4f}')
    print(f'  Total coherence = {winner["total_coherence"]:.1f}')
    print(f'  Grace efficiency = {winner["efficiency"]:.1f}%')
    verdict_09 = 'PASS' if winner['name'] == 'Biblical' else 'PARTIAL'
else:
    verdict_09 = 'FAIL'

ALL_RESULTS['09'] = {'verdict': verdict_09, 'winner': winner['name'] if valid else 'NONE'}

# Comparison bar chart
fig, ax = plt.subplots(figsize=(12, 6))
names = [r['name'] for r in results_09]
coherences = [r['total_coherence'] for r in results_09]
colors = [COLORS[1] if r['ok'] else COLORS[2] for r in results_09]
bars = ax.bar(names, coherences, color=colors)
ax.set_title(f'Total Coherence by Strategy — Winner: {winner["name"] if valid else "NONE"}', fontsize=14)
ax.set_ylabel('Total Coherence (integral of C over time)')
for bar, r in zip(bars, results_09):
    label = 'VALID' if r['ok'] else 'VIOLATED'
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, label,
            ha='center', va='bottom', fontsize=9, color='white')
plt.tight_layout()
plt.show()
print(f'\nVERDICT: {verdict_09}')
```
